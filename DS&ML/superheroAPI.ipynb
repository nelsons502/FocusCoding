```python
import requests
import json
import pandas as pd
import re

token = "1c5fe6e4750a253c7ae3f6ed8e2f9111"

heros = pd.DataFrame(columns=['name', 'id', 'intelligence', 'strength', 'speed', 'durability', 'power', 'combat'])

i = 1  # Start with hero ID 1
while True:
    url = f'https://www.superheroapi.com/api/{token}/{i}'  # Fixed URL format
    response = requests.get(url)

    if response.status_code != 200:
        print(f"Error fetching hero {i}, status code: {response.status_code}")
        break  # Exit loop if the API call fails



    # Extract JSON from response text
    match = re.search(r'({.*})', response.text, re.DOTALL)  # Look for JSON object
    if match:
        clean_json = match.group(1)  # Extract just the JSON part
        hero = json.loads(clean_json)  # Manually parse JSON
    else:
        print(f"Failed to extract JSON for hero {i}")
        print("Raw response:", response.text)
        exit()
    

    
    if 'powerstats' not in hero:  # Handle missing stats
        print(f"Skipping hero {i}, missing powerstats")
        i += 1
        continue


    # Use `.loc[len(heros)]` to add a new row properly
    heros.loc[len(heros)] = [
        hero.get('name', 'Unknown'),
        hero.get('id', 'Unknown'),
        hero['powerstats'].get('intelligence', 'Unknown'),
        hero['powerstats'].get('strength', 'Unknown'),
        hero['powerstats'].get('speed', 'Unknown'),
        hero['powerstats'].get('durability', 'Unknown'),
        hero['powerstats'].get('power', 'Unknown'),
        hero['powerstats'].get('combat', 'Unknown')
    ]

    i += 1  # Move to the next hero

    
display(heros)
```
